```python
#import matplotlib, cartopy.crs and cartopy.feature
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature

#create figure
fig, ax = plt.subplots(subplot_kw={'projection': ccrs.SouthPolarStereo()})

#set the extent 
ax.set_extent([-180, 180, -90, -60], ccrs.PlateCarree())

#add features
ax.add_feature(cfeature.COASTLINE, edgecolor='black')
ax.add_feature(cfeature.LAND, color='lightgray')
ax.add_feature(cfeature.OCEAN, color='lightblue')
ax.gridlines(draw_labels=True)

#add title
ax.set_title('Antarctic Continent and Southern Ocean', fontsize=14)

```

```python
#import matplotlib, cartopy.crs, cartopy.feature, numpy and Nominatim
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import numpy as np
from geopy.geocoders import Nominatim

#set the extent
extent = [-80, 30, -50, -10]
central_lon = np.mean(extent[:2])
central_lat = np.mean(extent[2:])

#set the locations
geolocator = Nominatim(user_agent="my_geocoder")
place = ['Walvis Bay','Cape Town','Rio de Janeiro','Montevideo']
address = []
for p in place:
    loc = geolocator.geocode(p,language="en")
    address.append(loc)
print(address)

#create figure
plt.figure(figsize=(12, 6))
ax = plt.axes(projection=ccrs.AlbersEqualArea(central_lon, central_lat))
ax.set_extent(extent)

#add features
ax.add_feature(cfeature.OCEAN, color='lightblue')
ax.gridlines(draw_labels=True)
for p in range(len(place)):
    ax.text(address[p].longitude,address[p].latitude,place[p],transform=ccrs.Geodetic())
```

```python
#import matplotlib.pyplot, cartopy.crs, cartopy.feature and numpy
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import numpy as np

#set the resolutions
cl_coarse = cfeature.GSHHSFeature(scale='c')
cl_intermediate = cfeature.GSHHSFeature(scale='i')
cl_full = cfeature.GSHHSFeature(scale='f')

#set the extent
central_lon, central_lat = 18, -34.2
extent = [17.5, 19.5, -34, -34.5]

#create figures and add features
fig,ax = plt.subplots(1,3,figsize=(20,18),subplot_kw={'projection':ccrs.Orthographic(central_lon, central_lat)})
ax[0].set_extent(extent)
gl0 = ax[0].gridlines(draw_labels=True)
gl0.right_labels = False
gl0.top_labels = False
ax[0].add_feature(cfeature.COASTLINE, edgecolor='black')
ax[0].add_feature(cfeature.LAND, color='lightgray')
ax[0].add_feature(cfeature.OCEAN, color='lightblue')
ax[0].add_feature(cl_coarse)
ax[0].set_title('False Bay Coarse Resolution', fontsize=12)

ax[1].set_extent(extent)
gl1 = ax[1].gridlines(draw_labels=True)
gl1.right_labels = False
gl1.top_labels = False
ax[1].add_feature(cfeature.COASTLINE, edgecolor='black')
ax[1].add_feature(cfeature.LAND, color='lightgray')
ax[1].add_feature(cfeature.OCEAN, color='lightblue')
ax[1].add_feature(cl_intermediate)
ax[1].set_title('False Bay Intermediate Resolution', fontsize=12)

ax[2].set_extent(extent)
gl2 = ax[2].gridlines(draw_labels=True)
gl2.right_labels = False
gl2.top_labels = False
ax[2].add_feature(cfeature.COASTLINE, edgecolor='black')
ax[2].add_feature(cfeature.LAND, color='lightgray')
ax[2].add_feature(cfeature.OCEAN, color='lightblue')
ax[2].add_feature(cl_full)
ax[2].set_title('False Bay Full Resolution', fontsize=12)

plt.show()
```
